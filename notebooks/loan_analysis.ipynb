```python
#importing the libraries and loading the data

import pandas as pd
import numpy as np

#load the data

df = pd.read_csv('../data/bank_loan_raw.csv')
```

```python
#checking the no of rows and columns in the data
df.shape
#checking the dataset info
df.info()
```

```python
#checking the datatypes 
df.dtypes
```

```python
#checking for the missing values in the dataset
df.isnull().sum()

df['Gender'].value_counts()

df['Married'].value_counts()

df['Credit_History'].value_counts()

```

```python
# Fill categorical missing values using mode

df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['Married'].fillna(df['Married'].mode()[0], inplace=True)
df['Self_Employed'].fillna(df['Self_Employed'].mode()[0], inplace=True)

#checking the values after filling the missing values
df[['Gender','Married','Self_Employed']].isnull().sum()

```

```python
#checking for the dependent 
df['Dependents'].value_counts()

#replace 3+ with 3
df['Dependents']=df['Dependents'].replace('3+',3)

#converting the dependents column to numeric
df['Dependents']=pd.to_numeric(df['Dependents'])

#checking the datatyppe after converting only for specific dependent column
df['Dependents'].dtypes

#replacing the missing values in the dependents column with the mode value
df['Dependents'].fillna(df['Dependents'].mode()[0],inplace=True)

#checking the values after filling the missing values
df['Dependents'].isnull().sum()
```

```python
#loanamaount loanamount term and credit history
#checking the null values for loanamount and loan amount term and credit history
df[['LoanAmount','Loan_Amount_Term','Credit_History']].isnull().sum()

df['LoanAmount'].describe()

#checcking for right skewness in the loan amount column
df['LoanAmount'].median(), df['LoanAmount'].mean()

#filling the missing values in the loan amount column with the median value
df['LoanAmount'].fillna(df['LoanAmount'].median(),inplace=True)

#checking the values after filling the missing values
df['LoanAmount'].isnull().sum()

#checking the values in the loan amount term column
df['Loan_Amount_Term'].value_counts()

#filling the missing values in the loan amount term column with the mode value
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0],inplace=True)

#checcking the values after filling the missing values
df['Loan_Amount_Term'].isnull().sum()

#checking the values in the credit history column
df['Credit_History'].value_counts()

#filling the missing values in the credit history column with the mode value
df['Credit_History'].fillna(df['Credit_History'].mode()[0],inplace=True)

#checking the values after filling the missing values
df['Credit_History'].isnull().sum()
```

```python
#final check for the missing values in the dataset
df.isnull().sum()

#checking for the dupliactes
df.duplicated().sum()

#creating the total income column by adding the applicant income and coapplicant income
df['Total_Income']=df['ApplicantIncome']+df['CoapplicantIncome']
#creating the income loan ratio column by dividing the total income by the loan amount
df['Income_Loan_Ratio']=df['Total_Income']/df['LoanAmount']

#exporting the cleaned data to a new csv file
df.to_csv('../cleaned_data/bank_loan_cleaned.csv',index=False)
```
